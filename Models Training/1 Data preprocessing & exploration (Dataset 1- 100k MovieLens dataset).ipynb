<h1>1.1 Initial data preprocessing & exploration (Dataset 1- 100K MovieLens Dataset)</h1>

<h2>1.1.1 Install dependencies</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from zipfile import ZipFile
import zipfile
from torch.utils.data import Dataset
import torch
import os
import re
import requests
import sklearn
import random
import seaborn as sns
from sklearn.model_selection import StratifiedShuffleSplit
import warnings
warnings.filterwarnings("ignore")
```

<h2>1.1.2 Data Preprocessing & Splitting</h2>

```python
class DataFetcher:
    def __init__(self, directory: str, archive_name: str = 'ml-latest-small.zip', fetch: bool = False) -> None:
        self.directory = directory
        self.archive_name = archive_name
        self.archive_path = os.path.join(self.directory, self.archive_name)
        
        if fetch or not os.path.isfile(self.archive_path):
            self._fetch_dataset()
        
        self.user_ratings, self.film_details = self._unpack_and_preprocess()

    def _fetch_dataset(self) -> None:
        """
        Retrieves the dataset from a predefined source if it is not present locally.
        """
        url = "http://files.grouplens.org/datasets/movielens/ml-latest-small.zip" # 100k movielens dataset
        print('Retrieving dataset...')
        response = requests.get(url, stream=True)
        os.makedirs(self.directory, exist_ok=True)
        
        with open(self.archive_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print('Dataset retrieved!\n')
    
    def _unpack_and_preprocess(self):
        """
        Unpacks the dataset, loads user ratings and movie details into Pandas DataFrames, and preprocesses the datasets.
        """
        print("Unpacking dataset...")
        
        with zipfile.ZipFile(self.archive_path, 'r') as zip_ref:
            file_list = zip_ref.namelist()
            print("Contained files:", file_list)

            subfolder = 'ml-latest-small/'
    
            # Find files in the subfolder
            ratings_csv = next((f for f in file_list if 'ratings.csv' in f and f.startswith(subfolder)), None)
            movies_csv = next((f for f in file_list if 'movies.csv' in f and f.startswith(subfolder)), None)
            
            if not ratings_csv or not movies_csv:
                raise FileNotFoundError("Required CSV files not found in the subfolder!")
        
            ratings_df = pd.read_csv(zip_ref.open(ratings_csv))
            ratings_df['timestamp'] = pd.to_datetime(ratings_df['timestamp'], unit='s')
            
            movies_df = pd.read_csv(zip_ref.open(movies_csv))

            # One-hot encoding
            genre_expanded = movies_df.genres.str.get_dummies(sep='|')
            genre_tally = genre_expanded.sum(axis=1)
            movies_df = pd.concat([movies_df.iloc[:,:-1], genre_expanded, genre_tally.rename('genreTally')], axis=1)

            # Remove movies without any associated ratings
            movies_df = movies_df[movies_df.movieId.isin(ratings_df.movieId)].reset_index(drop=True)

            # Remove ratings linked to movies with no genre
            ratings_df = ratings_df[ratings_df.movieId.isin(movies_df[movies_df['(no genres listed)'] == 0].movieId)]
            
            # Extract year from movie title and add as a new feature
            movie_years = movies_df.title.apply(lambda x: re.findall('\(([0-9]*)\)\s*$', x)[0] if len(re.findall('\(([0-9]*)\)\s*$', x)) > 0 else '9999')
            movies_df['releaseYear'] = movie_years.astype('int64')

            # Remove ratings linked to movies with the year marked as 9999
            ratings_df = ratings_df[ratings_df.movieId.isin(movies_df[movies_df.releaseYear != 9999].movieId)]
            movies_df = movies_df[movies_df.releaseYear != 9999][movies_df['(no genres listed)'] == 0].reset_index(drop=True)
            movies_df.drop(columns={'(no genres listed)'}, inplace=True)
            
            cols = movies_df.columns.tolist()
            cols.remove('releaseYear')
            cols.insert(2, 'releaseYear')
            movies_df = movies_df[cols]

            ratings_df = pd.merge(ratings_df, movies_df, how='inner', on='movieId')

        print("Unpacking complete!\n")
        return ratings_df, movies_df

    def train_test_split(self):
        """
        Divides the user ratings into training and test sets.
        Leave-One-Out Methodology: For each user, the model will be tested on the most recently rated movie and trained on all of the movies rated 
                                before the most recent.
        """
        print("Train-Test splitting...")
        
        # Assign recency ranking to each user's rating
        self.user_ratings['latest'] = self.user_ratings.groupby(['userId'])['timestamp'].rank(method='first', ascending=False)
        
        # Generate test dataset (most recent rating per user) - 
        test_set = self.user_ratings[self.user_ratings.latest == 1][['userId', 'movieId', 'rating', 'timestamp']]
        
        # Generate train dataset (all other ratings)
        train_set = self.user_ratings[self.user_ratings.latest != 1][['userId', 'movieId', 'rating', 'timestamp']]
        
        # Convert ratings into implicit feedback
        train_set.loc[:, 'rating'] = 1
        test_set.loc[:, 'rating'] = 1

        test_set = pd.merge(test_set, self.film_details, how='inner', on='movieId')
        
        print(f"Rating Total: {len(self.user_ratings)}, Movie Total: {len(self.film_details)}, Train: {len(train_set)}, Test: {len(test_set)}")
        print("Train-Test splitting complete!\n")
        return self.user_ratings, self.film_details, train_set, test_set


# Instantiate the class and get the data
data_loader = DataFetcher(directory='Datasets')
ratings_data, movies_data, train_data, test_data = data_loader.train_test_split()

print("Preprocessed files downloading")

# preprocessed movies csv
movies_data.to_csv("./Datasets/movies_preprocessed_D1.csv", index = False)

# preprocessed ratings csv
ratings_data.to_csv("./Datasets/ratings_preprocessed_D1.csv", index = False)

# test csv
test_data.to_csv("./Datasets/test_D1.csv", index = False)

# train csv
train_data.to_csv("./Datasets/train_D1.csv", index = False)

print("Contained files: ['movies_preprocessed_D1.csv', 'ratings_preprocessed_D1.csv', 'test_D1.csv', 'train_D1.csv']")
print("Preprocessed files download complete!\n")
```

<h2>1.1.3 Exploratory Data Analysis</h2>

<h3>1.1.3.1 Distribution of Movie Ratings</h3>

```python
rating_counts = ratings_data['rating'].value_counts().sort_index()

ax = rating_counts.plot(kind='bar', figsize=(10, 6), rot=0, color='skyblue')
plt.title('Distribution of Movie Ratings', pad=20)
plt.xlabel('Rating Value')
plt.ylabel('Count of Ratings')

# Add percentage labels
total = rating_counts.sum()
for p in ax.patches:
    ax.annotate(f'{p.get_height()/total:.1%}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), 
                textcoords='offset points')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

<h3>1.1.3.2 Distribution of User Average Ratings</h3>

```python
# Calculate average rating per user
avg_ratings_per_user = ratings_data.groupby('userId')['rating'].mean()

# Create styled histogram
plt.figure(figsize=(12, 6))
ax = sns.histplot(avg_ratings_per_user,
                 bins=20,
                 kde=True,
                 color='dodgerblue',  # Pure blue color
                 edgecolor='white',
                 linewidth=0.5,
                 alpha=0.8)

# Add statistical markers
mean_val = avg_ratings_per_user.mean()
median_val = avg_ratings_per_user.median()
plt.axvline(mean_val, color='navy', linestyle='-', linewidth=1.5, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='red', linestyle='--', linewidth=1.5, label=f'Median: {median_val:.2f}')

# Add legend and labels
plt.legend(loc='upper right')
plt.title('Distribution of User Average Ratings', fontsize=14, pad=15)
plt.xlabel('Average Rating Given', fontsize=12)
plt.ylabel('Number of Users', fontsize=12)

# Enhance readability
plt.grid(axis='y', linestyle=':', alpha=0.4)
sns.despine()
plt.xlim(0, 5)  # For 5-star rating systems
plt.tight_layout()

plt.show()
```

<h3>1.1.3.3 Top 10 most rated movies</h3>

```python
# Count ratings per movie and merge with titles
most_rated = ratings_data.groupby('movieId').size().reset_index(name='num_ratings')
most_rated = most_rated.merge(movies_data[['movieId', 'title']], on='movieId')
top_10_most_rated = most_rated.nlargest(10, 'num_ratings')

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=top_10_most_rated, y='title', x='num_ratings', palette='viridis')
plt.xlabel('Number of Ratings')
plt.ylabel('Movie Title')
plt.title('Top 10 Most Rated Movies')

# Add data labels
for i, count in enumerate(top_10_most_rated['num_ratings']):
    plt.text(count + 5, i, str(count), va='center')

plt.tight_layout()
plt.show()
```

<h3>1.1.3.4 Top 10 Movies by 5-Star Ratings</h3>

```python
# Filter ratings where rating is 5
top_ratings = ratings_data[ratings_data['rating'] == 5]

# Count number of users who rated each movie 5 stars
movie_counts = top_ratings.groupby('movieId').size().reset_index(name='num_users')

# Merge with movie titles and release year
top_movies = movie_counts.merge(movies_data[['movieId', 'title', 'releaseYear']], on='movieId')

# Create a combined title with year for better readability
top_movies['title_with_year'] = top_movies['title'] + ' (' + top_movies['releaseYear'].astype(str) + ')'

# Get top 10 most user-rated 5-star movies
top_10_movies = top_movies.nlargest(10, 'num_users').sort_values('num_users', ascending=True)

# Plot with enhanced styling
plt.figure(figsize=(14, 8))

# Create horizontal bar plot with highest at top
bars = plt.barh(top_10_movies['title'], 
                top_10_movies['num_users'],
                color=plt.cm.plasma_r(np.linspace(0.2, 0.8, 10)),  # Reverse color gradient
                height=0.7,
                edgecolor='white',
                linewidth=0.5)

# Add value labels 
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.01*max(top_10_movies['num_users']),  # Position outside bar
             bar.get_y() + bar.get_height()/2,
             f'{int(width)}',
             va='center',
             ha='left',
             color='black',
             fontsize=11)

# Formatting
plt.title('Top 10 Movies by 5-Star Ratings\n', fontsize=16, pad=20)
plt.xlabel('\nNumber of 5-Star Ratings', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)
plt.xlim(0, max(top_10_movies['num_users']) * 1.15)  # Add padding

# Customize ticks and grid
plt.xticks(fontsize=10)
plt.yticks(fontsize=11)
plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

<h3>1.1.3.8 Top 10 Highest Rated Movies with more than 10 ratings</h3>

```python
# The Mean Rating for Each Movie
mean_ratings = ratings_data.groupby(['movieId'])['rating'].mean().round(2).reset_index().merge(movies_data.iloc[:,:2], on='movieId')
mean_ratings.columns = ['movieId', 'mean_rating', 'title']

count_ratings = ratings_data.movieId.value_counts().reset_index()
count_ratings.columns = ['movieId', 'count_rating']

ratings_aggr = mean_ratings.merge(count_ratings, on='movieId')

# Filter for movies with >10 ratings
filtered_ratings = ratings_aggr[ratings_aggr['count_rating'] > 10]

# Create the distribution plot
plt.figure(figsize=(12, 6))
ax = sns.histplot(filtered_ratings['mean_rating'], 
                 bins=30,
                 kde=True,
                 color='teal',
                 edgecolor='white')

# Add median line
median = filtered_ratings['mean_rating'].median()
plt.axvline(median, color='red', linestyle='--', linewidth=1)
plt.text(median+0.05, plt.ylim()[1]*0.9, f'Median: {median:.2f}', color='red')

plt.title('Distribution of Mean Movie Ratings\n(Movies with >10 ratings)', fontsize=14)
plt.xlabel('Average Rating')
plt.ylabel('Number of Movies')
plt.grid(axis='y', alpha=0.3)
sns.despine()
plt.show()
```

<h3>1.1.3.9 Top 10 Highest Rated Movies with more than 10 ratings</h3>

```python
# Count ratings per movie
count_ratings = ratings_data['movieId'].value_counts().reset_index()
count_ratings.columns = ['movieId', 'count_rating']

# Merge and filter
ratings_aggr = mean_ratings.merge(count_ratings, on='movieId')
filtered_ratings = ratings_aggr[ratings_aggr['count_rating'] > 10]
top_movies = filtered_ratings.nlargest(10, 'mean_rating').sort_values('mean_rating', ascending=True)

# Create the top movies plot
plt.figure(figsize=(12, 8))
bars = plt.barh(top_movies['title'], 
                top_movies['mean_rating'],
                color=plt.cm.viridis(np.linspace(0.2, 0.8, 20)),
                edgecolor='white',
                height=0.7)

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width - 0.05, 
             bar.get_y() + bar.get_height()/2,
             f'{width:.2f}',
             va='center',
             ha='right',
             color='white',
             fontweight='bold')

plt.title('Top 20 Highest Rated Movies\n(Minimum 10 ratings)', fontsize=14)
plt.xlabel('Average Rating')
plt.ylabel('Movie Title')
plt.xlim(0, 5.2)
plt.grid(axis='x', linestyle=':', alpha=0.3)
sns.despine(left=True, bottom=True)
plt.show()
```
